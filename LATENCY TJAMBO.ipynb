```python
import pandas as pd
import json
import numpy as np
import pingparsing
```

```python
city =['Dallas','Denver','Miami','Chicago','Minneapolis','Montreal','New York','San Francisco','Seattle','Washington DC','Bogota','Buenos Aires','Amsterdam','Copenhagen','Frankfurt']
```

```python
IP = ["69.162.81.155","192.199.248.75","162.254.206.227","209.142.68.29","207.250.234.100","184.107.126.165","206.71.50.230","65.49.22.66","23.81.0.59","207.228.238.7","200.7.98.19","131.255.7.26","95.142.107.181","185.206.224.67","195.201.213.247"] 
```

```python

ping_parser = pingparsing.PingParsing()
transmitter = pingparsing.PingTransmitter()

Latency = [] 

for address in IP : 
    transmitter.destination = address
    transmitter.count = 1
    result = transmitter.ping()
    response =json.dumps(ping_parser.parse(result).as_dict(), indent=4)
    res = json.loads(response)
    Latency.append(res['rtt_max'])

```
